## 5-VAZIFA: Countplot() orqali Attrition ustunidagi ma'lumotlarni "Yes" "No" soniga ko'ra, ustunli grafik shaklida tasvirlang

<a href="https://imgbb.com/"><img src="https://i.ibb.co/Px9Bn6J/plt.png" alt="plt" border="0"></a>

```python
# JAVOBNI SHUYERGA YOZING
attrition_counts = data['Attrition'].value_counts()
# Grafikni ko'rsatish
attrition_counts.plot(kind='bar')
# Grafikni joylash va nomlash
plt.title("Attrition Distribution")
plt.xlabel("Attrition")
plt.ylabel("Count")

# Grafikni ko'rsatish
plt.show()
```

## 6-VAZIFA: Countplot() orqali Attrition ustunidagi ma'lumotlarni "Yes" "No" soniga ko'ra hamda gender ustuniga bo'lingan holda ustunli grafikda tasvirlang.

<a href="https://imgbb.com/"><img src="https://i.ibb.co/R7BC8z2/bygender.png" alt="bygender" border="0"></a>

```python
# JAVOBNI SHUYERGA YOZING
# "Attrition" va "Gender" ustunlariga qarab ma'lumotlarni kesib olish
attrition_gender_data = data[['Attrition', 'Gender']]
# Kesib olingan ma'lumotlarni "Attrition" va "Gender" bo'yicha guruhlash
attrition_gender_counts = attrition_gender_data.groupby(['Attrition', 'Gender']).size().unstack()
# Grafikni ko'rsatish
attrition_gender_counts.plot(kind='bar', stacked=True)
# Grafikni joylash va nomlash
plt.title("Attrition Distribution by Gender")
plt.xlabel("Attrition")
plt.ylabel("Count")

# Grafikni ko'rsatish
plt.show()
```

```python
# "Attrition" va "Gender" ustunlariga qarab ma'lumotlarni kesib olish
attrition_gender_data = data[['Attrition', 'Gender']]

# Kesib olingan ma'lumotlarni "Attrition" va "Gender" bo'yicha guruhlash
attrition_gender_counts = attrition_gender_data.groupby(['Attrition', 'Gender']).size().unstack()

# Grafiklarni alohida oynalab chiqarish
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# "Yes" va "No" qiymatlarni alohida oynalash
attrition_gender_counts.loc['Yes'].plot(kind='bar', ax=axes[0], title='Attrition: Yes')
attrition_gender_counts.loc['No'].plot(kind='bar', ax=axes[1], title='Attrition: No')

# Umumiy oynalashni yaxshilash
plt.tight_layout()

# Grafiklarni ko'rsatish
plt.show()
```

## 7-VAZIFA: Countplot() orqali Attrition ustunidagi ma'lumotlarni "Yes" "No" soniga ko'ra hamda jobLevel ustuniga bo'lingan holda ustunli grafikda tasvirlang.

<a href="https://imgbb.com/"><img src="https://i.ibb.co/kmzVfDK/byjoblevel.png" alt="byjoblevel" border="0"></a><br /><a target='_blank' href='https://the-crosswordsolver.com/tobagos-island-neighbor-8-letters'></a><br />

```python
# JAVOBNI SHUYERGA YOZING
# "Attrition" va "JobLevel" ustunlariga qarab ma'lumotlarni kesib olish
attrition_joblevel_data = data[['Attrition', 'JobLevel']]

# "Attrition" va "JobLevel" ustunlariga qarab guruhlayib sonlarini hisoblash
attrition_joblevel_counts = attrition_joblevel_data.groupby(['Attrition', 'JobLevel']).size().unstack()

# Grafikni ko'rsatish
attrition_joblevel_counts.plot(kind='bar', stacked=True)

# Grafikni joylash va nomlash
plt.title("Attrition Distribution by Job Level")
plt.xlabel("Attrition")
plt.ylabel("Count")

# Grafikni ko'rsatish
plt.show()
```

## 8-VAZIFA: Pie chart orqali "Attrition" (ishdan bo'shaganmi) ustunidagi Yes va No qiymatlariga ko'ra grafikni chiqaring

<a href="https://imgbb.com/"><img src="https://i.ibb.co/LZ4cCKb/PieChart.png" alt="PieChart" border="0"></a><br /><a target='_blank' href='https://the-crosswordsolver.com/netman-smith-4-letters'></a><br />

```python
# JAVOBNI SHUYERGA YOZING
# "Attrition" ustunidagi qiymatlar
attrition_counts = data['Attrition'].value_counts()
# Pie chartni yaratish
plt.figure(figsize=(6, 6))
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#99ff99'])
# Diagramma nomi
plt.title("Attrition Distribution")
# Pie chartni ko'rsatish
plt.axis('equal')
plt.show()
```

## 9-VAZIFA:  Crosstab orqali "Attrition" (ishdan bo'shaganmi) ustunidagi Yes va No qiymatlari hamda Department ustuniga ko'ra grafikni chiqaring

<a href="https://imgbb.com/"><img src="https://i.ibb.co/g4KgG6q/depart.png" alt="depart" border="0"></a>

```python
# JAVOBNI SHUYERGA YOZING
# Ma'lumotlarni tuzatish
# Crosstab yordamida ma'lumotlarni to'plash
attrition_department_crosstab = pd.crosstab(data['Attrition'], data['Department'])

# Pie chartni yaratish
fig, ax = plt.subplots(figsize=(8, 6))
attrition_department_crosstab.plot(kind='pie', subplots=True, ax=ax, autopct='%1.1f%%', colors=['#66b3ff', '#99ff99', '#ffcc99'])

# Diagramma nomi
plt.title("Attrition Distribution by Department")

# Pie chartni ko'rsatish
plt.axis('equal')
plt.show()
```

```python
# Crosstab yordamida ma'lumotlarni to'plash
attrition_department_crosstab = pd.crosstab(data['Department'], data['Attrition'])

# Ustunli grafikni yaratish
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(data=attrition_department_crosstab, x=attrition_department_crosstab.index, y='Yes', color='blue', label='Yes')
sns.barplot(data=attrition_department_crosstab, x=attrition_department_crosstab.index, y='No', color='green', bottom=attrition_department_crosstab['Yes'], label='No')

# Grafikni joylash va nomlash
plt.title("Attrition Distribution by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.legend()

# Grafikni ko'rsatish
plt.show()
```

```python
# Ma'lumotlarni DataFrame ko'rinishida tuzish
df = pd.DataFrame(data)

# Crosstab yordamida ma'lumotlarni to'plash
crosstab_data = pd.crosstab(df['Department'], df['Attrition'])

# Ustunli grafikni yaratish
crosstab_data.plot(kind='bar', stacked=True)
```

```python
# Crosstab yordamida ma'lumotlarni to'plash
crosstab_data = pd.crosstab(data['Department'], data['Attrition'])

# Ustunli grafikni yaratish
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

# "Yes" ustuni alohida ko'rsatiladi
sns.barplot(data=crosstab_data, x=crosstab_data.index, y='Yes', color='blue', label='Yes')

# "No" ustuni "Yes" ustuni orqasidan ko'rsatiladi
sns.barplot(data=crosstab_data, x=crosstab_data.index, y='No', color='green', bottom=crosstab_data['Yes'], label='No')

# Grafikni joylash va nomlash
plt.title("Attrition Distribution by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.legend()

# Grafikni ko'rsatish
plt.show()
```

## 10-VAZIFA:  Seaborn kutubxonasining Distribution(taqsimot) imkoniyatidan foydalanib, Attrition(ishdan bo'shaganmi) ustunini MonthlyIncome ustuniga solishtirilganligini grafigini ekranga chiqaring.


**Hint: ** Bu orqali biz ishchining oylik maoshini uning ishdan bo'shab ketishiga bog'liq yoki bog'liq emasligini topamiz.

<a href="https://imgbb.com/"><img src="https://i.ibb.co/zQc4FVZ/displot.png" alt="displot" border="0"></a><br /><a target='_blank' href='https://500pxdownload.com/'></a><br />

```python
# JAVOBNI SHUYERGA YOZING

# "Attrition" ustunidagi "Yes" qiymatlarni 1 ga, "No" qiymatlarni 0 ga o'zgartirish
data['Attrition'] = data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Seaborn kutubxonasining distplot() yordamida grafikni chiqarish
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.distplot(data[data['Attrition'] == 0]['MonthlyIncome'], label='No Attrition', color='blue')
sns.distplot(data[data['Attrition'] == 1]['MonthlyIncome'], label='Attrition', color='red')

# Grafikni joylash va nomlash
plt.title("Distribution of Monthly Income by Attrition")
plt.xlabel("Monthly Income")
plt.ylabel("Density")
plt.legend()

# Grafikni ko'rsatish
plt.show()
```
