```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_excel('/content/Книга1.xls')
df.columns=['Methanol consumption', 'Air consumption','Air mixture concentration']
df.head(33)
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=df)
plt.title("AR( MC model qiymatlari chizmasi)")
plt.xlabel("vaqt")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.regplot(data=df, x='Methanol consumption', y='Air consumption')
plt.show()
```
